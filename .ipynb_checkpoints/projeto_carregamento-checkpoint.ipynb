```python
import fpdf
import webbrowser
import os
import pandas as pd
```

```python
class Cliente:
    def __init__(self,umaData, umNomeCliente,umCpf,umaUnidade,umValorAnterior, umChequeRecebido):
        self.data = umaData
        self.nomeCliente = umNomeCliente
        self.cpf = umCpf
        self.unidade = umaUnidade
        self.valorAnterior = umValorAnterior
        self.chequeRecebido = umChequeRecebido        
        self.valorCarregamentoDia = 0
        self.qtdNf = 0
        self.valorTotal = 0
    
    def alterarCarregamentoDia(self, alterarCarregamentoDia):
        self.valorCarregamentoDia = alterarCarregamentoDia
        
    def alterarNf(self, alterarNf):
        self.qtdNf = alterarNf
        
    def calcularValor(self):
        self.valorTotal = self.chequeRecebido + self.valorAnterior - self.valorCarregamentoDia
        
    def imprimir(self):
        print("Relatório de Crédito/Débitos pendentes", "\nUnidade:", self.unidade, "\nNome:", self.nomeCliente, "\nCPF:", self.cpf ,
             "\nData:", self.data, "\nCrédito/Débito anterior: R$", self.valorAnterior, "\nValor Recebido: R$", self.chequeRecebido, "\nNotas Fiscais do dia:", self.qtdNf,
              "\nValor total dos carregamentos do dia: R$", self.valorCarregamentoDia, "\nValor final: R$", self.valorTotal)        
        
        
        
def gerarNf():
    dicionario1 = {
        "Notas Fiscais": pd.Series(listaNf),
        "Valores": pd.Series(listaCarregamentoDia),
    }
    df = pd.DataFrame(dicionario1)
    return df

def gerarPdf(df):
    pdf = fpdf.FPDF()
    pdf.add_page()
    pdf.ln(5)
    pdf.image('logo_grupo_tubarao.png', x=(210-100)/2, y=8, w=95)
    pdf.ln(18)
    pdf.set_font("Arial", "B", size = 12)
    pdf.cell(200, 10, txt = f"Extrato de Carregamento", ln = True, align = 'C')
    pdf.ln(3)
    pdf.cell(200, 10, txt = f"Unidade {unidade}", ln = True, align = 'C')

    pdf.ln(10)
    pdf.set_font("Arial", size = 10)
    pdf.cell(200, 10, txt = f"Cliente: {nome}", ln = True, align = 'L')
    pdf.cell(200, 10, txt = f"CPF: {cpf}", ln = True, align = 'L')
    pdf.cell(200, 10, txt = f"Data do carregamento: {data}", ln = True, align = 'L')
    pdf.cell(200, 10, txt = f"Saldo anterior ao carregamento: R$ {valorAnterior}", ln = True, align = 'L')
    pdf.cell(200, 10, txt = f"Valor recebido no dia do carregamento: R$ {chequeRecebido}", ln = True, align = 'L')
    pdf.ln(7)
    pdf.set_font("Arial","U",size = 10)
    pdf.cell(200, 10, txt = f"Detalhes do Carregamento:", ln = True, align = 'L')        
    pdf.ln(2)
    pdf.set_font("Arial", size = 10)
    # Calcula a largura da tabela
    table_width = 100
    # Centraliza a tabela
    pdf.set_x((120 - table_width) / 2)
    pdf.cell(table_width / 2, 10, txt = "   Notas Fiscais   ", border = 1, align = 'C')
    pdf.cell(table_width / 2, 10, txt = "   Valores   ", border = 1, align = 'C')
    pdf.ln(10)
    for index, row in df.iterrows():
        pdf.set_x((120 - table_width) / 2)  # Centraliza cada linha da tabela
        pdf.cell(table_width / 2, 10, txt = str(row["Notas Fiscais"]), border = 1, align = 'C')
        pdf.cell(table_width / 2, 10, txt = str(row["Valores"]), border = 1, align = 'C')
        pdf.ln(10)
    pdf.ln(2)
    pdf.cell(200, 10, txt = f"Soma dos carregamentos: R${soma:.2f}", ln = True, align = 'L')
    pdf.set_font("Arial", "B", size = 10)
    pdf.cell(200, 10, txt = f"Saldo atual: R$ {cliente.valorTotal:.2f}", ln = True, align = 'L')
    pdf.set_font("Arial", size = 10)
    pdf.ln(10)
    pdf.cell(200, 10, txt=f"______________________________", align="C")
    pdf.ln(5)  # Adiciona uma linha em branco
    pdf.cell(200, 10, txt=f"{nome}", align="C")
    pdf.ln(10)  # Adiciona uma linha em branco
    pdf.cell(200, 10, txt=f"______________________________", align="C")
    pdf.ln(5)  # Adiciona uma linha em branco
    pdf.cell(200, 10, txt=f"Assinatura do Colaborador e Carímbo", align="C")

    
    
    pdf.set_y(-45)  # Posiciona o cursor 45 mm acima do final da página
    pdf.set_font("Arial", size=8)
    pdf.cell(0, 10, txt="WhatsApp Expedições: (12)99627-4813 - Aoki Filial | (12) 99715-3147 - Maçaim | (12) 99155-7775 - Tubarão 4", align="C")
    pdf.ln(5)  # Adiciona uma linha em branco
    pdf.cell(0, 10, txt="WhatsApp Escritório: (12)99117-9937 - Matriz", align="C")
    pdf.ln(5)  # Adiciona uma linha em branco
    pdf.cell(0, 10, txt="E-mail: comercial@areiastubarao.com.br", align="C")

    
    pdf.output("relatorio.pdf")
            
    # Abrir o arquivo PDF no navegador padrão
    webbrowser.open('file://' + os.path.realpath("Relatorio.pdf"))

```

```python
print("Bem vindo ao Sistema de Crédito/Débito de Carregamentos.")

listaCarregamentoDia = []
listaNf = []

while True:
    
    while True:
        try:
            validacao = int(input("Deseja realizar o relatório de Crédito/Débito? 1 - Sim/ 2 - Não\n"))
            
            if validacao not in [1, 2]:
                raise ValueError("Opção inválida. Por favor, escolha 1 ou 2.")
            break
        except ValueError as e:
            print(f"Erro: {e}")
    
    if validacao == 1:
        unidade = input("Digite a unidade do carregamento:\n")
        nome = input("Digite o nome do cliente:\n")
        cpf = input("Digite o CPF do cliente:\n")
        data = input("Digite a data do carregamento: (ex: 00/00/0000)\n")
        
        while True:
            try:
                valorAnterior = float(input("Digite o Crédito/Débito anterior ao carregamento:\n"))
                chequeRecebido = float(input("Digite o valor do cheque recebido:\n"))
                break
            except ValueError:
                print("Erro: Valor inválido. Por favor, digite um número.")
        
        cliente = Cliente(data,nome,cpf,unidade,valorAnterior,chequeRecebido)

        while True:
            try:                
                listaNf.append(input("Digite o código da Nota Fiscal do primeiro carregamento:\n"))
                listaCarregamentoDia.append(float(input("Digite o valor do primeiro carregamento:\n")))
                listaNf.append(input("Digite o código da Nota Fiscal do segundo carregamento:\n"))
                listaCarregamentoDia.append(float(input("Digite o valor do segundo carregamento:\n")))

                break
            except ValueError:
                print("Erro: Valor inválido. Por favor, digite um número.")

        while True:
            try:
                validacao_2 = int(input("Deseja acrescentar mais carregamentos? 1 - Sim/ 2 - Não\n"))
                
                if validacao_2 not in [1, 2]:
                    raise ValueError("Opção inválida. Por favor, escolha 1 ou 2.")
                break
            except ValueError as e:
                print(f"Erro: {e}")

        if validacao_2 == 1:
            while True:
                try:                    
                    listaNf.append(input("Digite o código da Nota Fiscal do terceiro carregamento:\n"))
                    listaCarregamentoDia.append(float(input("Digite o valor do terceiro carregamento:\n")))
                    listaNf.append(input("Digite o código da Nota Fiscal do quarto carregamento:\n"))
                    listaCarregamentoDia.append(float(input("Digite o valor do quarto carregamento:\n")))

                    break
                    
                except ValueError:
                    print("Erro: Valor inválido. Por favor, digite um número.")
            
            soma = sum(listaCarregamentoDia)
            nf = listaNf
            cliente.alterarCarregamentoDia(soma)
            cliente.calcularValor()
            cliente.imprimir()
            df = gerarNf()
            gerarPdf(df)
            
            break

        else:            
            nf = listaNf
            soma = sum(listaCarregamentoDia)
            cliente.alterarCarregamentoDia(soma)
            cliente.calcularValor()
            cliente.imprimir()
            df = gerarNf()
            gerarPdf(df)            
            break
            
```
