```python
import fpdf
import webbrowser
import os
import pandas as pd
import locale

class Cliente:
    def __init__(self,umaData, umNomeCliente,umCpf,umaUnidade,umValorAnterior, umChequeRecebido):
        self.data = umaData
        self.nomeCliente = umNomeCliente
        self.cpf = umCpf
        self.unidade = umaUnidade
        self.valorAnterior = umValorAnterior
        self.chequeRecebido = umChequeRecebido        
        self.valorCarregamentoDia = 0
        self.qtdNf = 0
        self.valorTotal = 0
    
    def alterarCarregamentoDia(self, alterarCarregamentoDia):
        self.valorCarregamentoDia = alterarCarregamentoDia
        
    def alterarNf(self, alterarNf):
        self.qtdNf = alterarNf
        
    def calcularValor(self):
        self.valorTotal = self.chequeRecebido + self.valorAnterior - self.valorCarregamentoDia
        
    def imprimir(self):
        print("Relatório de Crédito/Débitos pendentes", "\nUnidade:", self.unidade, "\nNome:", self.nomeCliente, "\nCPF:", self.cpf ,
              "\nData:", self.data, "\nCrédito/Débito anterior: R$", self.valorAnterior, "\nValor Recebido: R$", self.chequeRecebido,
              "\nValor total dos carregamentos do dia: R$", self.valorCarregamentoDia, "\nValor final: R$", self.valorTotal)        
        
def gerarNf(carregamento_Dia):
    notas_fiscais = [nota["Nota Fiscal"] for nota in carregamento_Dia]
    valores = [nota["Valor"] for nota in carregamento_Dia]
    
    dicionario1 = {
        "Notas Fiscais": notas_fiscais,
        "Valores": valores
    }
    df = pd.DataFrame(dicionario1)
    return df

def gerarPdf(df):
    
    locale.setlocale(locale.LC_ALL, 'pt_BR')

    pdf = fpdf.FPDF()
    pdf.add_page()
    pdf.set_margins(10, 20, 10)
    
    # Defina a funcao para o cabeçalho
    def header(pdf):
        pdf.image('logo_grupo_tubarao.png', x=(250-100)/2, y=8, w=65)
        pdf.ln(15)
        pdf.set_font("Helvetica", "B", size = 10)
        pdf.cell(200, 10, txt = f"RELATÓRIO DE CARREGAMENTO", ln = True, align = 'C')
        pdf.cell(200, 10, txt = f"Unidade - {unidade}", ln = True, align = 'C')
        pdf.ln(5)
        
    # Defina a função para o rodape    
    def footer(pdf):
        pdf.set_y(-42)  # Move to 1.5 cm from bottom
        pdf.set_font("Helvetica", size=8)
        pdf.cell(0, 10, txt="WhatsApp Expedições: (12) 99627-4813 - Aoki Filial | (12) 99715-3147 - Maçaim | (12) 99155-7775 - Tubarão 4", align="C")
        pdf.ln(5)  # Adiciona uma linha em branco
        pdf.cell(0, 10, txt="WhatsApp Escritório: (12) 99117-9937 - Matriz", align="C")
        pdf.ln(5)  # Adiciona uma linha em branco
        pdf.cell(0, 10, txt="E-mail: comercial@areiastubarao.com.br", align="C")


    header(pdf)    
    pdf.set_font("Helvetica", "B", size = 10)
    pdf.cell(15, 10, txt = "Cliente:", ln = False, align = 'L')
    pdf.set_font("Helvetica", size = 10)
    pdf.cell(0, 10, txt = f"{nome}", ln = True, align = 'L')
    pdf.set_font("Helvetica", "B", size = 10)
    pdf.cell(10, 10, txt = "Data:", ln = False, align = 'L')
    pdf.set_font("Helvetica", size = 10)
    pdf.cell(0, 10, txt = f"{data}", ln = True, align = 'L')    
    pdf.set_font("Helvetica", "B", size = 10)
    pdf.cell(9, 10, txt = "CPF:", ln = False, align = 'L')
    pdf.set_font("Helvetica", size = 10)
    pdf.cell(0, 10, txt = f"{cpf}", ln = True, align = 'L')
    pdf.set_font("Helvetica", style="B")  # Define a fonte em negrito
    pdf.cell(56, 10, txt = "Saldo anterior ao carregamento:", ln = False, align = 'L')
    pdf.set_font("Helvetica")  # Volta a definir a fonte regular
    pdf.cell(0, 10, txt = f"R$ {locale.format_string('%.2f', valorAnterior, grouping=True)}", ln = True, align = 'L')
    pdf.set_font("Helvetica", style="B")  # Define a fonte em negrito
    pdf.cell(31, 10, txt = "Valor depositado:", ln = False, align = 'L')
    pdf.set_font("Helvetica")  # Volta a definir a fonte regular
    pdf.cell(0, 10, txt = f"R$ {locale.format_string('%.2f', chequeRecebido, grouping=True)}", ln = True, align = 'L')
    pdf.ln(7)
    pdf.set_font("Helvetica","B",size = 10)
    pdf.cell(200, 10, txt = f"Detalhes dos Carregamentos:", ln = True, align = 'L')
    pdf.ln(2)
    pdf.set_font("Helvetica", size = 10)
    
    # Tabela
    table_width = 130    
    pdf.set_x((150 - table_width) / 2)
    pdf.set_font("Helvetica", "B", size = 10)
    pdf.cell(table_width / 2, 10, txt = "   Nota Fiscal   ", border = 1, align = 'C')
    pdf.cell(table_width / 2, 10, txt = "   Valor   ", border = 1, align = 'C')
    pdf.set_font("Helvetica", size = 10)
    pdf.ln(10)
    
    for index, row in df.iterrows():
        pdf.set_x((150 - table_width) / 2)  # Centraliza cada linha da tabela
        pdf.cell(table_width / 2, 10, txt = str(row["Notas Fiscais"]), border = 1, align = 'C')
        pdf.cell(table_width / 2, 10, txt = locale.currency(row["Valores"]), border = 1, align = 'C')
        pdf.ln(10)
        
    pdf.ln(2)
    pdf.set_font("Helvetica", "B", size = 10)
    pdf.cell(19, 10, txt = f"Valor total:", ln = False, align = 'L')
    pdf.set_font("Helvetica", size = 10)
    pdf.cell(0, 10, txt = f"R$ {locale.format_string('%.2f', soma_valores, grouping=True)}", ln = True, align = 'L')
    pdf.set_font("Helvetica", "B", size = 10)
    pdf.ln(5)
    pdf.cell(200, 10, txt = f"Saldo atual: R$ {locale.format_string('%.2f', cliente.valorTotal, grouping=True)}", ln = True, align = 'L')
    pdf.ln(7)
    pdf.cell(200, 10, txt=f"______________________________", align="C")
    pdf.ln(5)  # Adiciona uma linha em branco
    pdf.cell(200, 10, txt=f"{nome}", align="C")
    pdf.ln(17)  # Adiciona uma linha em branco
    pdf.cell(200, 10, txt=f"______________________________", align="C")
    pdf.ln(5)  # Adiciona uma linha em branco
    pdf.cell(200, 10, txt=f"Assinatura e Carímbo", align="C")
    
    footer(pdf)   
    
    pdf.output(f"carregamento_{nome}.pdf")
            
    # PDF ira abrir no navegador padrao
    webbrowser.open('file://' + os.path.realpath(f"carregamento_{nome}.pdf"))    
    
 
    
    
#MAIN --------------------------------------------------MAIN

print("Bem vindo ao Sistema de Crédito/Débito de Carregamentos.")

acesso = True
carregamento_Dia = []

while True:
    
    while True:
        try:
            validacao = int(input("Realizar o relatório de carregamento? 1 - Sim/ 2 - Não\n"))
            
            if validacao not in [1, 2]:
                raise ValueError("Opção inválida. Por favor, escolha 1 ou 2.")
            break
            
        except ValueError as e:
            print(f"Erro: Valor inválido. Por favor, digite um número.")
    
    if validacao == 1:
        
        unidade = input("Digite a unidade do carregamento:\n")
        nome = input("Digite o nome do cliente:\n")
        cpf = input("Digite o CPF do cliente:\n")
        data = input("Digite a data do carregamento: (ex: 00/00/0000)\n")
        
        while True:
            try:
                valorAnterior = float(input("Digite o saldo anterior ao carregamento:\n"))
                chequeRecebido = float(input("Digite o valor depositado:\n"))
                break

            except ValueError:
                print("Erro: Valor inválido. Por favor, digite um número.")

        cliente = Cliente(data,nome,cpf,unidade,valorAnterior,chequeRecebido)
        
        while acesso == True:
            try:
                entrada = int(input("Adicionar carregamento? - 1 - Sim/ 2 - Não\n"))

                if entrada == 1:
                    try:
                        num_nf = input("Digite o número da Nota Fiscal:\n")
                        valor_nf = float(input("Digite o valor da Nota Fiscal:\n"))

                        if valor_nf <= 0:
                            raise ValueError("Valor da Nota Fiscal deve ser maior que zero")

                        nota_fiscal = {"Nota Fiscal": num_nf, "Valor": valor_nf}
                        carregamento_Dia.append(nota_fiscal)

                    except ValueError as e:
                        print(f"Erro: Valor inválido. Por favor, digite um número")
                        continue

                elif entrada == 2:
                    if not carregamento_Dia:
                        print("Erro: Não há carregamentos para processar.")
                        continue

                    soma_valores = 0

                    for nota in carregamento_Dia:
                        soma_valores += nota["Valor"]

                    cliente.alterarCarregamentoDia(soma_valores)
                    cliente.calcularValor()
                    df = gerarNf(carregamento_Dia)
                    cliente.imprimir()
                    acesso = False

                else:
                    print("Opção inválida. Por favor, digite 1 ou 2.")
                    continue

            except ValueError as e:
                print(f"Opção inválida. Por favor, digite 1 ou 2")
                continue

            except Exception as e:
                print(f"Erro: Valor inválido. Por favor, digite um número")
                continue

    else:
        print("Sistema finalizado!")
        break



    while True:
        try:
            validacao_1 = int(input("Gerar PDF? 1 - Sim/ 2 - Não\n"))

            if validacao_1 not in [1, 2]:
                raise ValueError("Opção inválida. Por favor, escolha 1 ou 2.")
            break
        except ValueError as e:
            print(f"Erro: Valor inválido. Por favor, digite um número.")

    if validacao_1 == 1:
        gerarPdf(df)
        print("Sistema finalizado!")
        break
    else:
        print("Sistema Finalizado!")
        break
```
